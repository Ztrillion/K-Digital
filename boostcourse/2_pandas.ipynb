# loc, iloc
- loc = 인덱스의 이름으로 특정위치의 값에 접근(로우값을 가져옴)
- iloc = 인덱스의 위치를 기준으로 칼럼값 가져옴

```python
s=pd.Series(np.nan,index=[29,28,27,26,25,24,23,1,2,3,4,5,6,7])
s
```

```python
#loc는 인덱스의 이름에 접근 iloc는 인덱스의 위치에 접근
s.loc[3:],s.iloc[3:]
```

# map for series

```python
s1 = Series(np.arange(20))
s1.head()
```

```python
# lambda함수 적용
s1.map(lambda x:x**2).tail(5)
```

```python
#dict 타입
z = {1:'A',2:'B',3:'C',7:'Q'}
s1.map(z)
```

# apply
- map과 달리, series 전체(column)에 해당 함수를 적용
- 입력값이 series 데이터로 입력받아 handling 가능

```python
data.head()
```

```python
f = lambda x: x.max() - x.min()
data.apply(f)
```

```python
def f(x):
    return Series([x.min(),x.max()],index=['min','max'])
data.apply(f)
```
